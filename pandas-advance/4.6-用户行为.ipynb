```python
# 
length = len(data)
df_churn = data[data['']=='Yes']
```

```python
def eda_calculate(column, types):
    """
    
    param @column: str
    param @types: list
    """
    print("\n==========", column, types)
    
    # 1.
    res_list = []
    rate_list = []
    for t in types:
        res = len(data[data[column]==t])
        # 3.
        rate = len(df_churn[df_churn[column]==t]) / res
        
        # 
        res_list.append({t: res}) 
        rate_list.append({t: rate}) 
    
    # 2.
    print("****************")
    for res in res_list: # res = { 3400}
        print(res, "", list(res.values())[0] / length)

    # 4.
    print("****************")
    for rate in rate_list:        
        before = list(rate.values())[0]
        index = rate_list.index(rate) + 1
        print(rate)
        
        if index < len(rate_list):
            after = list(rate_list[index].values())[0]
            print("", before / after)
```

```python
# 
for c in ['', '','']:
    eda_calculate(column=c, types=data[c].drop_duplicates().tolist())
```
