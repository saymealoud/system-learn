## 3.3.2 Resample函数

- 计数、均值、方差、累加、累乘
- 周期转换
- 数据验证：for循环 vs 内置函数

```python
import pandas as pd
import pandas_datareader as pdr
```

```python
data = pdr.get_data_fred('GS10')
data
```

```python
type(data)
```

```python
# 计算：均值、方差、最大值、最小值、计数等常见的描述性指标
type(data.index)
```

```python
data.resample?
```

```python
data.resample(rule='YE').mean()
```

```python
count = data.resample(rule='YE').count()
count
```

```python
sum1 = data.resample(rule='YE').sum()
sum1
```

```python
sum1 / count
```

```python
 data.resample(rule='YE').std() # 数据样本内部的差异性
```

```python
data.resample(rule='YE').std().max()
```

```python
data.resample(rule='YE').prod()
```

```python
# 加载新数据
data2 = pdr.get_data_fred('GS5')
data2
```

```python
data['GS5'] = data2 # 值和index的精确匹配
data
```

```python
data['mean'] = (data['GS10'] + data['GS5']) / 2
```

```python
data
```

```python
# 周期转换
data.resample(rule='Y')
```

```python
data.resample(rule='Y').last() # 查看每年最后1个月的数据
```
